```python
from sympy import *
x = Symbol('x')
fx = x**3-50*x
gx= -1*x**4 + 88*x**2-241
fgx = fx - gx
```

```python
rts = solve(fgx)
rts
```

```python
dict_ = {}
k = 1
for i in rts:
    dict_[k] = {'x': round(N(i), 5), 'y': round(N(-1*i**4 + 88*i**2-241), 5)}
    k+=1
```
